```python
import numpy as np
import plotly.graph_objects as go
import math

##coord_x = int(input("Coordenada en x"))
##coord_y = int(input("Coordenada en y"))
##coord_z = int(input("Coordenada en z"))


def reflect(I, n, calculate_angle_theta_1, hit_point, limit_x_positive):

    #module = math.sqrt((hit_point[0])**2 + (hit_point[1])**2 + (hit_point[2])**2)
    angulo_radianes = math.radians(calculate_angle_theta_1)
    theta_2 = math.degrees(angulo_radianes)
    #print(theta_2, 'theta 2')    

    
    z = limit_x_positive * math.sin(theta_2)
    y = limit_y_positive * math.cos(theta_2)
    
    return [0,y,z]


def calculate_angle_theta_1(hipot):
    #module_cat_op = math.sqrt( (cat_op[0])**2 + (cat_op[1])**2 + (cat_op[2])**2 )
    #module_cat_ad = math.sqrt( (cat_ad[0])**2 + (cat_ad[1])**2 + (cat_ad[2])**2 )

    
    
    module_hipot = math.sqrt( (hipot[0])**2 + (hipot[1])**2 + (hipot[2])**2 )
    arcose_radians = math.acos(hipot[1]/module_hipot)
    arcoseno_degrees = math.degrees(arcose_radians)
    #print(arcoseno_degrees, 'arcoseno_degrees')
    theta_1 = 180 - 90 - arcoseno_degrees

    #print(theta_1, 'theta 1')
    return theta_1
    
    #theta =  
    
    ## fx = |f|cosA
    ## cosA = fx / |f|
    ## A = arcos(fx / |f|)



    

def linea_normal(hit_point, limit_y_negative, limit_y_positive):
    list_point = []
    p1 = np.array([hit_point[0], hit_point[0]+limit_y_negative, hit_point[2]])
    p2 = np.array([hit_point[0], hit_point[0]+limit_y_positive, hit_point[2]])
    list_point.append(p1)
    list_point.append(p2)
    return list_point
    

def refract(I, n, n1, n2, calculate_angle_theta_1, hit_point):

    #Ley de snell
    angulo_radianes = math.radians(calculate_angle_theta_1)
    seno_value = math.sin(angulo_radianes)
    
    #print(seno_value)
    rho = (n1 * seno_value)/n2
    sin_radians = math.asin(rho)
    theta_3 = math.degrees(sin_radians)


    #print(hit_point, 'punto de choque')
    #print(theta_3, 'theta 3')

    #fx = |f|cosA
    #|f| = fx/cosA

    theta_3_radians = theta_3 * (math.pi/180)

    coord_fx = y2 - y1
    
    module_f = coord_fx/(math.cos(theta_3_radians))

    ##print(math.cos(theta_3), 'coseno de 28')

    #print(module_f, 'module_f')
    
    fy = module_f*math.sin(theta_3_radians)
    
    fy_pitagoras = math.sqrt( (module_f)**2 - (coord_fx)**2)

    #print(fy, 'fy')
    #print(fy_pitagoras, 'fy_pitagoras')
    
    return [0, coord_fx, fy ], False, theta_3


def second_refract(point, n1, n2, calculate_angle_theta_1, limit_x_positive, limit_y_positive, limit_z_positive, theta_3):
    theta_total_3 = 180 - 90 - ( 90 - theta_3)
    angulo_radianes = math.radians(theta_total_3)
    seno_value = math.sin(angulo_radianes)

    #print(seno_value, 'seno_value')
    rho = (n2 * seno_value)/n1
    #print(rho, 'rho')
    sin_radians = math.asin(rho)
    theta_4 = math.degrees(sin_radians)

    z = limit_x_positive * math.sin(theta_4)
    y = limit_x_positive * math.cos(theta_4)
    
    
    return [ [0,0], [point[1],  point[1] + y], [point[2], point[2]+z]] 

plane_y = 2
thickness = 3   

limit_y_negative = 0.1
limit_y_positive = 5

limit_x_negative = 0
limit_x_positive = 3

limit_z_negative = 0
limit_z_positive= 5

y1 = plane_y
y2 = thickness
plane_normal = np.array([0, 1, 0]) 

n1, n2 = 1.0, 1.5

origin = np.array([0, 0, 0])
direction = norm(np.array([0, 1, 1]))  # inclinación

#print(direction, 'inclinacion')

t = (y1 - origin[1]) / direction[1] #indificar a que plano chocara en ZX
hit_point = origin + t * direction
#hit_point = [1.5, 2, 2.5]


#print(hit_point, 'hit_point')

R = reflect(direction, plane_normal, calculate_angle_theta_1(hit_point), hit_point, limit_x_positive)

#print(R, 'R')

L_N = linea_normal(hit_point,limit_y_negative, limit_y_positive)


#print(hit_point, "hit point")
#print(L_N[1][1], "Linea de la normal")


T, tir, theta_3 = refract(direction, plane_normal, n1, n2, calculate_angle_theta_1(hit_point), hit_point)

ST = second_refract([ hit_point[0] + T[0], hit_point[1] + T[1], hit_point[2] + T[2]], n1, n2, calculate_angle_theta_1(hit_point), limit_x_positive, limit_y_positive, limit_z_positive, theta_3)
#print(ST, 'ST')
SLN = linea_normal(np.array([hit_point[0] + T[0], hit_point[1] + T[1], hit_point[2] + T[2]]),limit_y_negative, limit_y_positive)


fig = go.Figure()

## -----------Planos ZX--------------
xx_xz, zz_xz = np.meshgrid(np.linspace(0, 3, 10), np.linspace(0, limit_z_positive, 10))
yy_xz_1 = np.ones_like(xx_xz) * y1
yy_xz_2 = np.ones_like(xx_xz) * y2

fig.add_trace(go.Surface(x=xx_xz, y=yy_xz_1, z=zz_xz, colorscale=[[0,'gray'],[1,'gray']],opacity=0.4, showscale=False))
fig.add_trace(go.Surface(x=xx_xz, y=yy_xz_2, z=zz_xz,colorscale=[[0,'gray'],[1,'gray']],opacity=0.4, showscale=False))


##-------------Planos ZY--------------
zz_zy, yy_zy = np.meshgrid(np.linspace(0, limit_z_positive, 10), np.linspace(2, 3 , 10))
xx_zy_1 = np.ones_like(zz_zy) * limit_x_negative
xx_zy_2 = np.ones_like(zz_zy) * limit_x_positive

fig.add_trace(go.Surface(x=xx_zy_1, y=yy_zy, z=zz_zy, colorscale=[[0, 'gray'],[1,'gray']], opacity=0.4, showscale=False))
fig.add_trace(go.Surface(x=xx_zy_2, y=yy_zy, z=zz_zy, colorscale=[[0, 'gray'],[1,'gray']], opacity=0.4, showscale=False))


##-------------Planos XY------------
xx_xy, yy_xy = np.meshgrid(np.linspace(0, 3, 10), np.linspace(2, 3, 10))
zz_xy_1= np.ones_like(xx_xy) * limit_z_negative
zz_xy_2= np.ones_like(xx_xy) * limit_z_positive

fig.add_trace(go.Surface(x=xx_xy, y=yy_xy, z=zz_xy_1, colorscale=[[0, 'gray'],[1,'gray']], opacity=0.4, showscale=False))
fig.add_trace(go.Surface(x=xx_xy, y=yy_xy, z=zz_xy_2, colorscale=[[0, 'gray'],[1,'gray']], opacity=0.4, showscale=False))



## Rayos Trazados

fig.add_trace(go.Scatter3d(
    x=[L_N[0][0], L_N[1][0]],
    y=[L_N[0][1], L_N[1][1]],
    z=[L_N[0][2], L_N[1][2]],
    mode='lines',
    line=dict(color='orange', width=6, dash="dash"),
    marker=dict(size=3, color='orange'),
    name='Lineal normal 1'
))

fig.add_trace(go.Scatter3d(
    x=[SLN[0][0], SLN[1][0]],
    y=[SLN[0][1], SLN[1][1]],
    z=[SLN[0][2], SLN[1][2]],
    mode='lines',
    line=dict(color='orange', width=6, dash="dash"),
    marker=dict(size=3, color='orange'),
    name='Lineal normal 2'
))

fig.add_trace(go.Scatter3d(
    x=[origin[0], hit_point[0]],
    y=[origin[1], hit_point[1]],
    z=[origin[2], hit_point[2]],
    mode='lines+markers',
    line=dict(color='red', width=6),
    marker=dict(size=3, color='red'),
    name='Incidente'
))

fig.add_trace(go.Scatter3d(
    x=[hit_point[0], (hit_point[0] + R[0])],
    y=[hit_point[1], -(hit_point[0] + R[0])],
    z=[hit_point[2], (hit_point[2] + R[2])],
    mode='lines+markers',
    line=dict(color='blue', width=6, dash="dash"),
    marker=dict(size=3, color='blue'),
    name='Reflejado'
))

if not tir:
    fig.add_trace(go.Scatter3d(
        x=[hit_point[0], hit_point[0] + T[0]],
        y=[hit_point[1], hit_point[1] + T[1]],
        z=[hit_point[2], hit_point[2] + T[2]],
        mode='lines+markers',
        line=dict(color='green', width=6, dash="dot"),
        marker=dict(size=3, color='green'),
        name='Refractado'
    ))
    fig.add_trace(go.Scatter3d(
        x=[ST[0][0], ST[0][1] ],
        y=[ST[1][0], ST[1][1]],
        z=[ST[2][0], ST[1][1]],
        mode='lines+markers',
        line=dict(color='purple', width=6, dash="dot"),
        marker=dict(size=3, color='purple'),
        name='Segundo Refractado'
    ))
#print(hit_point[0], hit_point[0] + T[0], 'en x2')
#print(hit_point[1], hit_point[1] + T[1], 'en y2')
#print(hit_point[2], hit_point[2] + T[2], 'en z2')

#print(ST[0], 'en x')
#print(ST[1][0], 'en y')
#print(ST[2][1], 'en z')



fig.update_layout(
    scene=dict(
        xaxis=dict(title='X'),
        yaxis=dict(title='Y'),
        zaxis=dict(title='Z'),
        aspectmode="cube"
    ),
    width=700,
    height=700,
    title="Reflexión y Refracción en 3D"
)

fig.show()
```
